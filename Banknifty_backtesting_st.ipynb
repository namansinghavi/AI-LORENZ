```python
def get_indicators_and_positions_df(df):
    """ This Function takes one arguments
    'df' : (dataframe) Dataframe on which indicators are needed to be calculated
    """
    df['prev_close'] = df['close'].shift(1)
    df['prev_sup'] = df['supertrend'].shift(1)
    df['prev_prev_close'] = df['close'].shift(2)
    df['prev_prev_sup'] = df['supertrend'].shift(2)
    df.dropna(inplace=True)


    df.loc[(df['prev_prev_close'] < df['prev_prev_sup']) & (df['prev_close'] > df['prev_sup']) & (df['close'] > df['supertrend']), 'in_position'] = 'long'
    df.loc[(df['prev_prev_close'] > df['prev_prev_sup']) & (df['prev_close'] < df['prev_sup']) & (df['close'] < df['supertrend']), 'in_position'] = 'short'
    return df
```

```python
bnf_position_df = get_indicators_and_positions_df(df)
```

```python

def calculate_rsi_adx(data, rsi_period=14, adx_period=14):
    """
    Calculates the RSI, ADX, +DMI and -DMI values for the given OHLC data using the specified periods.
    :param ohlc_data: A pandas DataFrame containing OHLC data with columns 'open', 'high', 'low', and 'close'.
    :param rsi_period: The period to use for RSI calculation (default is 14).
    :param adx_period: The period to use for ADX calculation (default is 14).
    :return: A tuple containing four pandas Series: the RSI values, +DMI values, -DMI values, and ADX values.
    """
    # Calculate RSI
    data['rsi'] = ta.momentum.RSIIndicator(close=data['close'], n=rsi_period).rsi()

    # Calculate ADX, +DMI, and -DMI
    adx_dmi = ta.trend.ADXDMIIndicator(high=data['high'], low=data['low'], close=data['close'], n=adx_period)
    data['adx'] = adx_dmi.adx()
    data['+dmi'] = adx_dmi.adx_pos()
    data['-dmi'] = adx_dmi.adx_neg()

    # Return RSI, ADX, +DMI, and -DMI values as Series
    return data['rsi'], data['+dmi'], data['-dmi'], data['adx']
```
